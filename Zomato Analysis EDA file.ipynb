```python
df['cuisines_list'] = df['cuisines'].apply(lambda x: x.split(', '))

# Step 3: Flatten the list of cuisines to get all individual cuisine names
all_cuisines = [cuisine for sublist in df['cuisines_list'] for cuisine in sublist]

# Step 4: Count the occurrences of each cuisine
cuisine_counts = pd.Series(all_cuisines).value_counts()

# Display the most popular cuisines
print("Most Popular Cuisines:")
print(cuisine_counts)

# Step 5: Visualize the most popular cuisines
plt.figure(figsize=(12, 6))
sns.barplot(x=cuisine_counts.index[:10], y=cuisine_counts.values[:10], palette='coolwarm')
plt.title('Top 10 Most Popular Cuisines Among Restaurants')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants Offering the Cuisine')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 1: Calculate the number of cuisines each restaurant offers
df['num_cuisines'] = df['cuisines_list'].apply(len)

# Step 1: Compute the correlation between the number of cuisines and aggregate rating
correlation = df['num_cuisines'].corr(df['aggregate_rating'])

print("Correlation between the number of cuisines and restaurant ratings:", correlation)

# Step 3 : Visualize the correlation
plt.figure(figsize=(12, 6))
sns.scatterplot(x='num_cuisines', y='aggregate_rating', data=df, color='teal')
plt.title('Number of Cuisines vs Restaurant Rating')
plt.xlabel('Number of Cuisines Offered')
plt.ylabel('Restaurant Rating')
plt.show()
```

```python
# Step 1: Analyze the relationship between price range and restaurant ratings
# Group by 'price_range' and calculate average rating
price_range_rating = df.groupby('price_range')['aggregate_rating'].mean().reset_index()

# Step 2: Visualize average rating for different price ranges
plt.figure(figsize=(10, 6))
sns.barplot(x='price_range', y='aggregate_rating', data=price_range_rating, palette='coolwarm')
plt.title('Average Restaurant Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.show()
```

```python
price_range_cost = df.groupby('price_range')['average_cost_for_two'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='price_range', y='average_cost_for_two', data=price_range_cost, palette='coolwarm')
plt.title('Average Cost for Two People by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Cost for Two')
plt.show()
```

```python
# Convert 'online_order' to categorical with 'Yes' and 'No'
#df['delivery'] = df['delivery'].map({1: 'Yes', -1: 'No'})
#df['delivery'] = df['delivery'].replace({1: 'Yes', -1: 'No'})

# Calculate average rating for restaurants with and without online ordering
delivery_order_rating = df.groupby('delivery')['aggregate_rating'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='delivery', y='aggregate_rating', data=delivery_order_rating, palette='viridis')
plt.title('Average Restaurant Rating by Online Order Availability')
plt.xlabel('Online Order Availability')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Step 1: Group by restaurant names and count the number of outlets
outlet_counts = df.groupby('name').size().reset_index(name='Outlet_Count')

# Step 2: Sort by number of outlets and select the top 10 chains
top_chains = outlet_counts.sort_values(by='Outlet_Count', ascending=False).head(10)

# Step 3: Merge the top chains with the original dataset to get ratings
top_chains_ratings = pd.merge(df, top_chains, on='name')

# Step 4: Calculate the average rating for each top chain
avg_ratings = top_chains_ratings.groupby('name')['aggregate_rating'].mean().reset_index()

# Step 5: Merge the average ratings with outlet counts
top_chains_final = pd.merge(top_chains, avg_ratings, on='name')

# Step 6: Visualize the top chains by outlet count
plt.figure(figsize=(10, 6))
plt.bar(top_chains_final['name'], top_chains_final['Outlet_Count'], color='skyblue')
plt.xlabel('Restaurant Chain')
plt.ylabel('Number of Outlets')
plt.title('Top Restaurant Chains by Number of Outlets')
plt.xticks(rotation=45)
plt.show()

# Step 7: Visualize the average ratings of top chains
plt.figure(figsize=(10, 6))
plt.bar(top_chains_final['name'], top_chains_final['aggregate_rating'], color='orange')
plt.xlabel('Restaurant Chain')
plt.ylabel('Average Rating')
plt.title('Average Ratings of Top Restaurant Chains')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 1: Feature Extraction - Identify restaurants with Wi-Fi, Alcohol, etc.
# We'll create binary columns for each feature of interest
df['Wifi'] = df['highlights'].apply(lambda x: 1 if 'Wifi' in x else 0)
df['Alcohol'] = df['highlights'].apply(lambda x: 1 if 'Alcohol' in x else 0)

# Step 2: Feature Distribution - Calculate the percentage of restaurants with each feature
feature_distribution = df[['Wifi', 'Alcohol']].mean() * 100

# Visualize the feature distribution
plt.figure(figsize=(6, 4))
feature_distribution.plot(kind='bar', color='skyblue')
plt.title('Feature Distribution Across Restaurants')
plt.ylabel('Percentage of Restaurants')
plt.show()

# Step 3: Investigate correlation of features with ratings
# Group by feature presence (Wi-Fi, Alcohol) and calculate the average rating
wifi_rating = df.groupby('Wifi')['aggregate_rating'].mean().reset_index()
alcohol_rating = df.groupby('Alcohol')['aggregate_rating'].mean().reset_index()

# Step 4: Visualize the correlation between features and ratings
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
sns.barplot(x='Wifi', y='aggregate_rating', data=wifi_rating)
plt.title('Wifi Availability vs Average Rating')
plt.xlabel('Wifi Available')
plt.ylabel('Average Rating')

plt.subplot(1, 2, 2)
sns.barplot(x='Alcohol', y='aggregate_rating', data=alcohol_rating)
plt.title('Alcohol Availability vs Average Rating')
plt.xlabel('Alcohol Available')
plt.ylabel('Average Rating')

plt.tight_layout()
plt.show()
```
